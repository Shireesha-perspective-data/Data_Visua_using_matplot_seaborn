# why we use figure()
     we can use the plt.figure function to change the size of the figure.
if we want to change the size of the figure ,the figure might be too small and we want to increase the height of it,or we might want to increase the weight of it for that we can use the "plt.figure function" to change it's size.

```python
years=[2021,2022,2023,2024,2025,2026]
oranges=[0.34,0.12,0.12,0.9,0.5,0.3]
plt.figure(figsize=(5,8))  #here we wanto change the figure size
plt.plot(years,oranges,'^--r') #plt.plot(years,oranges,'or') here 'or' is the only points are ploted;if we wanto line in the formate fmt=[marker][line][color]
plt.title("yield of crop")  #title of the plot
plt.legend(["oranges"]) #here we wanto know the what represented line is that lebel;legend([lebel])
```

# why we use seaborn library
     already we are starting to draw prety good line graph,but one intresting thing that we can do just make a our charts a little more beautifull to the seaborn library.

```python
years=[2021,2022,2023,2024,2025,2026]
oranges=[0.34,0.12,0.12,0.9,0.5,0.3]
apple=[0.2,0.12,0.9,0.31,0.12,0.8] # style must be one of white, dark, whitegrid, darkgrid, ticks; box contain white 
sns.set_style('whitegrid') #here we wanto particular point is ploted.the grid of the exact point
plt.plot(years,oranges,'*-c',lw=5,ls=':') #[marker*][line--][colorblue],linewidth,linesyle means : are line form
plt.plot(years,apple,'or',ms='25',mec='b',mew=10,mfc='green',alpha=.10)

plt.xlabel("year")
plt.ylabel("fruits")

plt.title("crop yield in kanto")
plt.legend(['apple','oranges']);
```

**Format Strings**

A format string consists of a part for color, marker and line::

    fmt = '[marker][line][color]'

Each of them is optional. If not provided, the value from the style
cycle is used. Exception: If ``line`` is given, but no ``marker``,
the data will be a line without markers.

Other combinations such as ``[color][marker][line]`` are also
supported, but note that their parsing may be ambiguous.

**Markers**

=============   ===============================
character       description
=============   ===============================
``'.'``         point marker
``','``         pixel marker
``'o'``         circle marker
``'v'``         triangle_down marker
``'^'``         triangle_up marker
``'<'``         triangle_left marker
``'>'``         triangle_right marker
``'1'``         tri_down marker
``'2'``         tri_up marker
``'3'``         tri_left marker
``'4'``         tri_right marker
``'8'``         octagon marker
``'s'``         square marker
``'p'``         pentagon marker
``'P'``         plus (filled) marker
``'*'``         star marker
``'h'``         hexagon1 marker
``'H'``         hexagon2 marker
``'+'``         plus marker
``'x'``         x marker
``'X'``         x (filled) marker
``'D'``         diamond marker
``'d'``         thin_diamond marker
``'|'``         vline marker
``'_'``         hline marker
=============   ===============================

**Line Styles**

=============    ===============================
character        description
=============    ===============================
``'-'``          solid line style
``'--'``         dashed line style
``'-.'``         dash-dot line style
``':'``          dotted line style
=============    ===============================

Example format strings::

    'b'    # blue markers with default shape
    'or'   # red circles
    '-g'   # green solid line
    '--'   # dashed line with default color
    '^k:'  # black triangle_up markers connected by a dotted line

**Colors**

The supported color abbreviations are the single letter codes

=============    ===============================
character        color
=============    ===============================
``'b'``          blue
``'g'``          green
``'r'``          red
``'c'``          cyan
``'m'``          magenta
``'y'``          yellow
``'k'``          black
``'w'``          white
=============    ===============================
